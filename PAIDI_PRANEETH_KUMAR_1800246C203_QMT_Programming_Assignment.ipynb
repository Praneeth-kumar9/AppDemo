# 2) **Demonstrate a program for dot and tensor products of bra-and ket vectors.**

```python
import numpy as np
row_vec =np.array([2 , 5+6j ,5j ])
print ("bra vector is:",row_vec)
col_vec =np.array([[2] ,[5+6j] ,[5j] ])
print ("ket vector is:",col_vec)
```

```python
dp = np.dot(row_vec,col_vec)
print("dot product of two vectors",dp) 
tp = np.kron(row_vec,col_vec)
print("tensor product of two vectors",tp)
```

# 4) **Demonstrate the output of all three gates (X, Y, Z) connected in series for a single Q-bit system**

```python
qc=QuantumCircuit(1)
qc.x(0)
qc.y(0)
qc.z(0)
qc.draw()
```

```python
backend = Aer .get_backend('statevector_simulator')
final_state= execute(qc, backend).result().get_statevector() 
print(final_state)
```
